🎯 Want a quick map of what each module does?
Here’s a simple cheat sheet:

Module	                                                     Purpose
graph	                                             Build graphs (StateGraph, MessageGraph)
checkpoint	                                         Persistence + state saving
runtime	                                             Execution engine
channels	                                         Message passing between nodes
store	                                             Storage abstractions
types	                                             Core types used across LangGraph
constants	                                         START, END, etc.
managed	                                             Managed workflows
pregel	                                             Low-level graph execution model
